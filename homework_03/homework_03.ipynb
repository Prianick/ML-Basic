## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def convertStringStyle(text: str, format:str = '')-> str:
    text = text.replace(' ', '')

    if format != 'snake_case' and format != 'PascalCase':
        if text.find(('_')) != -1:
            format = 'snake_case'
        else:
            format = 'PascalCase'

    if format == 'snake_case':
        text = map(lambda x: x.capitalize(), text.split('_'))
    else:
        text = map(lambda x: '_' + x.lower() if x.isupper() else x, text)

    return ''.join(list(text)).strip('_')

text = input('Enter text: ')
format = input('Enter format (snake_case or PascalCase, leave empty for auto-detect): ')

print(convertStringStyle(text, format))
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
import re
from datetime import datetime

dateStr = input("Enter date in format DD.MM.YYYY: ")

def isLeapYear(year: int):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def isValidDate1(date: str) -> bool:
    if not re.match(r'^\d{2}\.\d{2}\.\d{4}$', date):
        return False

    parts = list(map(lambda x: int(x), date.split('.')))

    if parts[1] > 12:
        return False
    
    if parts[0] > 31:
        return False
    
    if (
        parts[0] > 30
        and parts[1] in (4,5,6,7,9,11)
    ):
        return False

    if (
        parts[1] == 2
        and (
            (
                isLeapYear(parts[2])
                and parts[0] > 29
            )
            or (
                not isLeapYear(parts[2])
                and parts[0] > 28
            )
        )
    ):
        return False
        
    return True

def isValidDate2(date_str):
    try:
        datetime.strptime(date_str, "%d.%m.%Y")
        return True
    except ValueError:
        return False

print(isValidDate1(dateStr))
#print(isValidDate2(dateStr))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def isPrime(number: int) -> bool:
    if number <= 1:
        return False
    if number in (2, 3):
        return True
    if number % 2 == 0:
        return False

    for divider in range(3, int(number**0.5) + 1, 2):
        if number % divider == 0:
            return False

    return True

print(isPrime(int(input())))
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
db = {}
longestNameLength = 0
longestSurnameLength = 0 

while True:
    info = input('Insert data. Format is "name surname age id".\n')

    if info == '':
        break

    parts = info.split()

    try:
        id = parts[3].zfill(8)
        name = parts[0].capitalize()
        if len(name) > longestNameLength:
            longestNameLength = len(name)
        surname = parts[1].capitalize()
        if len(surname) > longestSurnameLength:
            longestSurnameLength = len(surname)
        age = int(parts[2])

        if id in db:
            raise Exception('ID already exists.')

        if age < 18 or age > 60:
            raise Exception('Age is out of range')

        db[id] = (name, surname, str(age))

        print('Data was added.\n')
    except Exception as inst:
        print('Wrong data.')
        print(f"Error: {inst}\n")


for id in db:
    print(f"{id} {db[id][0]:{longestNameLength}} {db[id][1]:{longestSurnameLength}} {db[id][2]}")
```
